```python
import pandas as pd

t = pd.read_csv('/home1/yxiao977/proj/frip_sm_data/download_fastq/Yu_2022/SraRunTable.txt')
```

```python
t['genotype'] = t['genotype'].apply(lambda x : x.replace(" ", ""))
```

```python
t = t[t['Assay Type'] == 'ChIP-Seq']
chip = []
control = []
combine = []
for i, row in t.iterrows():
    chip.append(row['Run'])
    mask = mask = (
                    (t['treatment'] == row['treatment']) &
                    (t['genotype'] == row['genotype']) &
                    ((t['chip_antibody'].str.contains('input', case=False)) |
                    (t['chip_antibody'].str.contains('none', case=False))
                    )
                )
    ctrl = t[mask].iloc[0]
    control.append(ctrl['Run'])
    combine.append(row['chip_antibody'] + ' & ' + ctrl['chip_antibody'])
meta = pd.DataFrame({'ChIP': chip, 'Input': control, 'Combination':combine})
meta = meta[meta['Combination'] != 'none (input) & none (input)']
```
